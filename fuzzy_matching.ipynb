```python
import pandas as pd
```

```python
df_twitter = pd.read_excel("Filtered_Drive.xlsx", engine='openpyxl')
```

```python
df_nppes = pd.read_csv("Nppes data_gastro.csv")
```

```python
import pandas as pd
from fuzzywuzzy import process
import swifter

# Function to perform fuzzy matching
def fuzzy_match(name, choices):
    result = process.extractOne(name, choices)
    match, score = result[0], result[1]  # Unpacking only match and score
    with open("progress.log", "a") as f:
        f.write(f"{name} -> {match} ({score}%)\n")
    return pd.Series([match, score])

# Apply fuzzy matching
df_twitter[['Fullname', 'Fuzzy matching percentage']] = df_twitter['Name'].swifter.apply(
    lambda x: fuzzy_match(x, df_nppes['Fullname'])
)

df_twitter
```

```python

# Merge with df_nppes to get NPI
df_twitter_new = df_twitter.merge(df_nppes, on='Fullname', how='left')

# Reorder columns
df_twitter_new = df_twitter[['Name', 'Fullname', 'NPI', 'Link', 'Specialty', 'Address', 'Phone', 'Website', 'Page', 'Fuzzy matching percentage']]

df_twitter_new
```

```python
df_twitter_new.to_csv("FuzzyMatching.csv", encoding='utf-8')
```
