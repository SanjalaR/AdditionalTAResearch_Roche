```python
import pandas as pd
```

```python
df = pd.read_csv("with some columns - output.csv")
```

```python
import re
```

```python
def process_address(address):
    if type(address)==str:
        addr = [x.strip() for x in address.split(',')][-1]
        if addr!='':
            pattern = r"(.*) ([A-Z]{2}) (\d{5}(?:-\d{4})?)$"
            match = re.search(pattern, addr.strip())
            if match:
                city, state, postal_code = match.groups()
                #print('city: ', city, 'state: ', state,'code: ', postal_code)
                return ([city, state, postal_code])
    return ['', '', '']
```

```python
df['Address'] = df['Address'].str.rstrip(', ')
```

```python
new_cols = df['Address'].apply(process_address).apply(pd.Series)
```

```python
col_names = ['City', 'State', 'PostalCode']
new_cols.columns = col_names[:len(new_cols.columns)]
new_cols
```

```python
df1 = pd.concat([df, new_cols], axis=1)
```

```python
df1.to_csv('new new location with some cols - output.csv')
```
