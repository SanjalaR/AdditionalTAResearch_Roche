```python
import pandas as pd
from fuzzywuzzy import fuzz

def fuzzy_match(value1, value2):
    """Perform fuzzy matching between two values and return the match score."""
    if pd.isna(value1) or pd.isna(value2):  # Handle missing values
        return None, 0
    score = fuzz.ratio(str(value1), str(value2))  # Compute fuzzy match score
    return (value2, score) if score >= 80 else (None, score)  # Threshold set at 80

# Initialize new columns for matched values
df["matched_mod_postalcode"] = None
df["mod_postalcode_match_score"] = 0

# Perform fuzzy matching on the two postal code columns from the same row
for index, row in df.iterrows():
    best_postalcode, postalcode_score = fuzzy_match(
        row["PostalCode_crohnsandcolitis"], row["Provider Business Mailing Address Postal Code_nppes"]
    )

    df.at[index, "matched_mod_postalcode"] = best_postalcode
    df.at[index, "mod_postalcode_match_score"] = postalcode_score

# Display the updated DataFrame
print(df)
```

```python
import pandas as pd
from fuzzywuzzy import process

def fuzzy_match(value, choices, threshold=80):
    """Find the best fuzzy match for a given value from choices."""
    if not choices:  # If no choices available, return None
        return None, 0
    best_match, score = process.extractOne(value, choices)
    return (best_match, score) if score >= threshold else (None, score)

# Initialize new columns for matched values
df["matched_name_with_middlename"] = None
df["name_with_middlename_match_score"] = 0

# Perform fuzzy matching row-wise
for index, row in df.iterrows():
    best_name, name_score = fuzzy_match(
        row["Name_crohnsandcolitis"], [row["Fullname_with_middlename"]]  # Single-element list for row-wise matching
    )

    df.at[index, "matched_name_with_middlename"] = best_name
    df.at[index, "name_with_middlename_match_score"] = name_score

# Display the updated DataFrame
print(df)
```
